```python
import os
from pydub import AudioSegment

# Paths to the main folders
raw_folder = "/home/snicolau/bdb-classifier/bdb_dataset_raw"
processed_folder = "/home/snicolau/bdb-classifier/bdb_dataset"



for subfolder_name in os.listdir(raw_folder):
    if subfolder_name == "out":
        continue
    raw_subfolder_path = os.path.join(raw_folder, subfolder_name)
    processed_subfolder_path = os.path.join(processed_folder, subfolder_name)

    song_number = 1

    # Check if it's a directory
    if os.path.isdir(raw_subfolder_path):
        print(f"Processing subfolder: {subfolder_name}")

        # Create the corresponding processed subfolder if it doesn't exist
        os.makedirs(processed_subfolder_path, exist_ok=True)

        for file_name in os.listdir(raw_subfolder_path):
            raw_file_path = os.path.join(raw_subfolder_path, file_name)

            # Check if the file is an audio file (e.g., .mp3, .wav)
            if file_name.lower().endswith((".mp3", ".wav", ".flac", ".m4a")):
                print(f"Processing file: {file_name}")

                try:
                    # Load the audio file
                    audio = AudioSegment.from_file(raw_file_path)

                    # Calculate the number of 30-second fragments
                    fragment_duration_ms = 30 * 1000
                    num_fragments = len(audio) // fragment_duration_ms

                    for fragment_number in range(num_fragments):
                        # Extract the fragment
                        start_time = fragment_number * fragment_duration_ms
                        end_time = start_time + fragment_duration_ms
                        fragment = audio[start_time:end_time]

                        # Create a new file name
                        new_file_name = f"{song_number}_{fragment_number + 1}_{subfolder_name}.wav"
                        new_file_path = os.path.join(processed_subfolder_path, new_file_name)

                        # Export the fragment
                        fragment.export(new_file_path, format="wav")

                    # Increment the song number
                    song_number += 1

                except Exception as e:
                    print(f"Error processing file {file_name}: {e}")
```
